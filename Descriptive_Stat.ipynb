### Outlier Detection

Outlier Detection is the process of identifying data points that significantly differ from the rest of the dataset. Detecting outliers helps uncover anomalies, errors, or unusual observations that may affect data analysis.


### Outlier Detection using Z-Score

The Z-Score method identifies outliers by measuring how many standard deviations a data point is away from the mean. Typically, observations with a Z-Score greater than 3 or less than -3 are considered potential outliers.


```python
# OUTLIER DETECTION USING Z-SCORE

import pandas as pd
import numpy as np
from scipy.stats import zscore

# Sample dataset
data = {
    "ages": [23,24,32,45,12,43,67,45,32,56,32,120]
}

df = pd.DataFrame(data)

# Calculate Z-score
df["Z_Score"] = zscore(df["ages"])

# Threshold
threshold = 3

# Find outliers
outliers = df[np.abs(df["Z_Score"]) > threshold]

print("Dataset:")
print(df)

print("\nOutliers using Z-Score:")
print(outliers)
```

### Measures of Dispersion

Measures of Dispersion describe the spread or variability of a dataset. They indicate how far the data values are scattered from the central tendency, helping to understand the consistency and distribution of the data.
1.Variance
2.Standard Deviation


```python
statistics.variance(ages)
```

```python
np.var(ages,axis=0)
```

```python
def variance(data):
  n=len(ages)
  ## mean of the data
  mean=sum(data)/n
  ## variance
  deviation=[(x - mean) ** 2 for x in data]
  variance=sum(deviation)/(n-1)
  return variance
```

```python
variance(ages)
```

```python
def variance(data,dof=0):
  n=len(ages)
  ## mean of the data
  mean=sum(data)/n
  ## variance
  deviation=[(x - mean) ** 2 for x in data]
  variance=sum(deviation)/(n-dof)
  return variance
```

```python
variance(ages,dof=1)
```

### Normal Distribution

The Normal Distribution is a symmetric, bell-shaped probability distribution where most values cluster around the mean. It is one of the most widely used distributions in statistics and data science.

### Characteristics:

Symmetric bell-shaped curve
Mean = Median = Mode
Defined by mean (μ) and standard deviation (σ)


```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=0, scale=1)

plt.plot(x, y)
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()
```

### Log-Normal Distribution

A Log-Normal Distribution occurs when the logarithm of a variable follows a normal distribution. It is commonly used to model positively skewed data such as incomes, stock prices, and population sizes.

### Characteristics:

Right-skewed
Only positive values
Common in finance and biological measurements

```python
from scipy.stats import lognorm
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 10, 1000)
y = lognorm.pdf(x, s=0.5)

plt.plot(x, y)
plt.title("Log-Normal Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()
```

### Pareto Distribution (Power-Law Distribution)

The Pareto Distribution is a heavy-tailed distribution that follows the "80/20 rule," where a small proportion of observations accounts for a large proportion of the outcome. It is often used to model wealth, city sizes, and internet traffic.

### Characteristics:

Heavy-tailed
Demonstrates the 80-20 rule
Models wealth and resource distributions

```python
from scipy.stats import pareto
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 1000)
y = pareto.pdf(x, b=3)

plt.plot(x, y)
plt.title("Pareto Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()
```

### Covariance

Covariance measures the direction of the relationship between two variables. A positive value indicates that the variables tend to increase together, while a negative value indicates an inverse relationship.

### Correlation

Correlation measures the strength and direction of the relationship between two variables. Its value ranges from -1 to 1, where values closer to ±1 indicate a stronger relationship.

### Pearson Correlation

Pearson Correlation measures the strength and direction of a linear relationship between two continuous variables. It is widely used when the data is approximately normally distributed.

### Spearman Rank Correlation

Spearman Correlation measures the strength and direction of a monotonic relationship between two variables using their ranks. It is useful for non-linear relationships and ordinal data.
